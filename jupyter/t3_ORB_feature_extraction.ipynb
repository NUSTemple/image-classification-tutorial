```python
import warnings
warnings.simplefilter("ignore")
import numpy as np
import cv2
import os
import pandas as pd
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option('precision', 5)
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (10,10)

working_path = "C:\\Users\\pengtan\\Projects\\python\\image-classification-tutorial"
```

```python
oriimg = cv2.imread(os.path.join(working_path, "data", "wafer_map.jpg") ,0)
```

#### Original Image

```python
plt.imshow(oriimg,cmap='gray'),plt.show()
```

#### Resize Image to 128*128

```python
# img = cv2.resize(oriimg,None,fx=0.125,fy=0.125,interpolation=cv2.INTER_CUBIC)
img = cv2.resize(oriimg,(128,128),interpolation=cv2.INTER_AREA)
plt.imshow(img)
plt.show()
```

#### gaussian threshold

```python
# img = cv2.medianBlur(img, 3)
img= cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11,2)

plt.imshow(img)
plt.show()
```

```python
oriimg2 = cv2.imread(os.path.join(working_path, "data", "wafer_map.jpg"))
gray = cv2.cvtColor(oriimg2, cv2.COLOR_RGBA2GRAY)
edge=cv2.Canny(gray, 100, 150)
blank_img = oriimg2
blank_img[:,:] = 0

lines = cv2.HoughLinesP(edge, 1, np.pi / 360, 10, 3, 50)
for x in range(0, len(lines)):
    for x1,y1,x2,y2 in lines[x]:
        cv2.line(blank_img,(x1,y1),(x2,y2),(0,0,255), 3, cv2.LINE_AA)
        
plt.imshow(blank_img)
```

#### Calculate Image Gradient

```python
#compute gradient of image
oriimg = cv2.imread(os.path.join(working_path, "data", "wafer_map.jpg") ,0)
gx, gy = np.gradient(oriimg)

# plotting
plt.close("all")
plt.figure()
plt.suptitle("Image, and it gradient along each axis")
ax = plt.subplot("131")
ax.axis("off")
ax.imshow(img)
ax.set_title("image")

ax = plt.subplot("132")
ax.axis("off")
ax.imshow(gx,cmap='gray')
ax.set_title("gx")

ax = plt.subplot("133")
ax.axis("off")
ax.imshow(gy,cmap='gray')
ax.set_title("gy")
plt.show()
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=7)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=7)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
sift = cv2.xfeatures2d.SIFT_create()
surf = cv2.xfeatures2d.SURF_create()
orb = cv2.ORB_create(nfeatures=100)

keypoints_sift, descriptors = sift.detectAndCompute(img, None)
keypoints_surf, descriptors = surf.detectAndCompute(img, None)
keypoints_orb, descriptors = orb.detectAndCompute(img, None)

img_sift = cv2.drawKeypoints(img, keypoints_sift, None)
plt.imshow(img_sift),plt.show()

img_surf = cv2.drawKeypoints(img, keypoints_surf, None)
plt.imshow(img_surf),plt.show()

img_orb = cv2.drawKeypoints(img, keypoints_orb, None)
plt.imshow(img_orb),plt.show()
```

#### Transform the Image

```python
oriimg3 = cv2.imread(os.path.join(working_path, "data", "wafer_map.jpg"), 0)
radius = 950

polar_image = cv2.linearPolar(oriimg3,(oriimg3.shape[0]/2, oriimg3.shape[1]/2), 
                              radius, cv2.WARP_FILL_OUTLIERS)
polar_image = polar_image.astype(np.uint8)
plt.imshow(polar_image)


(h, w) = polar_image.shape[:2]
# calculate the center of the image
center = (w / 2, h / 2)
angle270 = 270
scale = 1.0
M = cv2.getRotationMatrix2D(center, angle270, scale)
rotated270 = cv2.warpAffine(polar_image, M, (h, w))
plt.imshow(rotated270)
plt.show()
```

```python
img = cv2.resize(rotated270,(256,256),interpolation=cv2.INTER_AREA)

sift = cv2.xfeatures2d.SIFT_create()
surf = cv2.xfeatures2d.SURF_create()
orb = cv2.ORB_create(nfeatures=100)

keypoints_sift, descriptors = sift.detectAndCompute(img, None)
keypoints_surf, descriptors = surf.detectAndCompute(img, None)
keypoints_orb, descriptors = orb.detectAndCompute(img, None)

img_sift = cv2.drawKeypoints(img, keypoints_sift, None)
plt.imshow(img_sift),plt.show()

img_surf = cv2.drawKeypoints(img, keypoints_surf, None)
plt.imshow(img_surf),plt.show()

img_orb = cv2.drawKeypoints(img, keypoints_orb, None)
plt.imshow(img_orb),plt.show()
```
